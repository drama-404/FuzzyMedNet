```python
# Import Necessary Libraries:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os
# import wandb
import warnings

from config import *
from src.data_preprocessing.demographics import *
from src.data_preprocessing.vitals_labs import *
from utils.safe_display import blind_display
```

```python
# Configuration & Settings:

# For better visualization
warnings.filterwarnings('ignore')
os.environ['WANDB_SILENT'] = 'true'
sns.set_theme(style="whitegrid")
```

```python
# Safety flag - when set to True will not display sensitive data
BLINDED = False
```

```python
# Data Loading
patients = pd.read_hdf(PROCESSED_DATA_FILE_PATH, 'patients')
vitals_labs = pd.read_hdf(PROCESSED_DATA_FILE_PATH, 'vitals_labs')
interventions = pd.read_hdf(PROCESSED_DATA_FILE_PATH, 'interventions')
```

```python
blind_display(patients, vitals_labs, interventions, blinded=BLINDED)
```

```python
X_patients = patients.drop(columns=['mort_hosp'])
y_patients = patients['mort_hosp']
```

# Filter Methods

## Variance Threshold

```python
X_patients_num = X_patients[['age', 'dnr_first', 'cmo_first', 'los_icu', 'readmission_30']]

from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.2)
selector.fit_transform(X_patients_num)
print(selector.get_support(indices=True))
```

```python
# Retrieve mean values per hour and flatter multi-index
idx = pd.IndexSlice
vitals_labs_mean = vitals_labs.loc[:, idx[:, 'mean']]
vitals_labs_mean.columns = [col[0] for col in vitals_labs_mean.columns.values]
blind_display(vitals_labs_mean, blinded=BLINDED)
```

```python
selector2 = VarianceThreshold(threshold=0.2)
selector2.fit_transform(vitals_labs_mean)
print(selector2.get_support(indices=True))
```

```python
selector3 = VarianceThreshold(threshold=0.2)
selector3.fit_transform(interventions)
print(selector3.get_support(indices=True))
```

## Pearson's Correlation

### Correlation between features

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_matrix = X_patients_num.corr(method='pearson') # 'pearson' is default

sns.heatmap(corr_matrix, annot=True, cmap='RdBu_r')
plt.show()
```

```python
corr_matrix2 = vitals_labs_mean.corr(method='pearson') # 'pearson' is default

fig, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(corr_matrix2, annot=False, cmap='RdBu_r')
plt.show()
```

```python
# Loop over bottom diagonal of correlation matrix
for i in range(len(corr_matrix2.columns)):
    for j in range(i):
        # Print variables with high correlation
        if abs(corr_matrix2.iloc[i, j]) > 0.7:
            print(corr_matrix2.columns[i] + ', ', corr_matrix2.columns[j] + ', ', corr_matrix2.iloc[i, j])
```

### Correlation between feature and target

```python
X_y_patients = X_patients_num.copy()
X_y_patients['mort_hosp'] = y_patients
```

```python
corr_matrix2 = X_y_patients.corr()

# Isolate the column corresponding to `exam_score`
corr_target = corr_matrix2[['mort_hosp']].drop(labels=['mort_hosp'])

sns.heatmap(corr_target, annot=True, fmt='.3', cmap='RdBu_r')
plt.show()
```

```python
X_y_vitals = pd.merge(vitals_labs_mean, patients['mort_hosp'], on=['subject_id', 'hadm_id', 'icustay_id'], how='left')
blind_display(X_y_vitals, blinded=BLINDED)
```

```python
corr_matrix3 = X_y_vitals.corr()

# Isolate the column corresponding to `exam_score`
corr_target = corr_matrix3[['mort_hosp']].drop(labels=['mort_hosp'])

fig, ax = plt.subplots(figsize=(8, 26))
sns.heatmap(corr_target, annot=True, fmt='.3', cmap='RdBu_r')
plt.show()
```
